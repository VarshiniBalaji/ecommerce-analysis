```python
def unique_values(column='abc'):
    df=pd.DataFrame()
    df=train_copy[column].value_counts().reset_index()
    if(len(df)>50):
        df.columns=[column,'count']
        df.to_csv('{}_count.csv'.format(column))
        print(column+'------Generated as csv')
        print(df.head(20))
    else:
        print(column+'unique value')
        print(df)
```

```python
def missing_data(df):
    df1=df.isna()
    for i in list(df.columns):
#         print(i)
#         print(df1[df1[i]==True])
        a=len(df1[df1[i]==True])
        if(a>0):
            print("the column with missing data",i)
            print("total missing data",len(df[df1[i]==True]))
            df[i]=df[i].fillna('noad')
            print(df)
    return df
```

```python
def encoding_label(column="abc"):
    if(column!='timestamp' and column!='checkout' and column!='order_placed'):
        try:
            var=0
            var = preprocessing.LabelEncoder()
            var.fit(train_copy[column])
            print(column)
            print(var.classes_)
            train_copy[column]=var.transform(train_copy[column])
            print(train_copy[column])
        except:
            print(column+' not done with label encoding - error')
    else:
        if(column=='timestamp'):
            print("timestamp is skipped")
        elif(column=='checkout'):
            print("checkout is not need")
        elif(column=='order_placed'):
            print("order_placed is not need")
```
